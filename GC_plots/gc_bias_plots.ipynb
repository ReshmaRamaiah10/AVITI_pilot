```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ch10_aviti = pd.read_csv('Ch10_scRNA_AVITI_gcbias.csv',header=None, names=["percentage","count"])
Ch10_illumina = pd.read_csv('Ch10_scRNA_Illumina_gcbias.csv',header=None, names=["percentage","count"])
Ch7_aviti = pd.read_csv('Ch7_scRNA_AVITI_gcbias.csv',header=None, names=["percentage","count"])
Ch7_illumina = pd.read_csv('Ch7_scRNA_Illumina_gcbias.csv',header=None, names=["percentage","count"])
```

```python
print("Total counts, Ch7 Illumina:",Ch7_illumina['count'].sum(),"AVITI:",Ch7_aviti['count'].sum())
print("Total counts, Ch10 Illumina:",Ch10_illumina['count'].sum(),"AVITI:",Ch10_aviti['count'].sum())
```

```python
# Ch7 GC bias plot (counts)
x = np.arange(100)
plt.plot(x, Ch7_aviti['count'], label='AVITI')
plt.plot(x, Ch7_illumina['count'], label='Illumina')
plt.xlabel('%GC')
plt.ylabel('Count')
plt.title("Ch7 GEX library (3') Per Sequence GC Content")
plt.legend()
plt.savefig('Ch7_scRNA_GCbias_count.png',dpi=500)
plt.show()
```

```python
# Ch10 GC bias plot (counts)
x = np.arange(100)
plt.plot(x, Ch10_aviti['count'], label='AVITI')
plt.plot(x, Ch10_illumina['count'], label='Illumina')
plt.xlabel('%GC')
plt.ylabel('Count')
plt.title("Ch10 GEX library (5') Per Sequence GC Content")
plt.legend()
plt.savefig('Ch10_scRNA_GCbias_count.png',dpi=500)
plt.show()
```

```python
# Ch7 GC bias plot (percentage)
x = np.arange(100)
plt.plot(x, Ch7_aviti['count']*100/Ch7_aviti['count'].sum(), label='AVITI')
plt.plot(x, Ch7_illumina['count']*100/Ch7_illumina['count'].sum(), label='Illumina')
plt.xlabel('%GC')
plt.ylabel('Percentage')
plt.title("Ch7 GEX library (3') Per Sequence GC Content (%)")
plt.legend()
plt.savefig('Ch7_scRNA_GCbias_percentage.png',dpi=500)
plt.show()
```

```python
# Ch10 GC bias plot (percentage)
x = np.arange(100)
plt.plot(x, Ch10_aviti['count']*100/Ch10_aviti['count'].sum(), label='AVITI')
plt.plot(x, Ch10_illumina['count']*100/Ch10_illumina['count'].sum(), label='Illumina')
plt.xlabel('%GC')
plt.ylabel('Percentage')
plt.title("Ch10 GEX library (5') Per Sequence GC Content (%)")
plt.legend()
plt.savefig('Ch10_scRNA_GCbias_percentage.png',dpi=500)
plt.show()
```
